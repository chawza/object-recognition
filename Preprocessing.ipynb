```python
%matplotlib inline
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bgr_image = cv2.imread('sample1.jpg') # https://www.waza.org/wp-content/uploads/2019/02/gsmp-amur-tiger-wz-m.jpg

rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
plt.title('our sample image')
plt.plot()
```

# Image colorspace

```python
plt.figure(figsize=(15,15))

plt.subplot(1,4,1)
plt.imshow(bgr_image[:,:,2], cmap="Reds")
plt.title('1 Channel color space of red')

plt.subplot(1,4,2)
plt.imshow(bgr_image[:,:,1], cmap="Greens")
plt.title('1 Channel color space of green')

plt.subplot(1,4,3)
plt.imshow(bgr_image[:,:,0], cmap='Blues')
plt.title('1 Channel color space of blue')

plt.subplot(1,4,4)
plt.imshow(rgb_image)
plt.title('3 Channel color space (RGB)')

plt.show()
```

# Grayscaling

```python
gray_img = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(15,15))

plt.subplot(1,2,1)
plt.imshow(rgb_image)
plt.title('3 Channel color space (RGB)')

plt.subplot(1,2,2)
plt.imshow(gray_img, cmap='gray')
plt.title('1 Channel color space (Grayscale)')

plt.plot()
```

# histogram equilization

```python
plt.figure(figsize=(15,15))
plt.subplot(2,2,1)
plt.imshow(gray_img, cmap='gray')
plt.title('original grayscale')

plt.subplot(2,2,2)
plt.hist(gray_img.flatten(), bins=256, range=(0,255))
plt.title('original intensity')

equ_img = cv2.equalizeHist(gray_img)

plt.subplot(2,2,3)
plt.imshow(equ_img, cmap='gray')
plt.title('equalized grayscale')

plt.subplot(2,2,4)
plt.hist(equ_img.flatten(), bins=256, range=(0,255))
plt.title('equalized intensity')

plt.show()
```
